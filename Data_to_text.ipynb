```python
###PASSSED

def extract_keyword(commentary_csv):
  Shots_name=['Straight drive','Cover drive','On drive','Square drive','Forward defence','sweep','reverse sweep','front foot leg glance','back foot drive','back foot defence','square cut','pull shot','hook shot','back foot leg glance','upper cut','helicopter shot','inside out shot','Marillier Shot','Switch Hit','Eoin Morgan Reverse Sweep','Dilscoop','Reverse Scoop','Ramp Shot','Periscope','Paddle Scoop']
  Ball_field=['Bowler','Wicket Keeper','Slip','Fly Slip','Gully','Third Man','Point','Sweeper','Long Stop','Cow Corner','Cover','Cover Point','Extra Cover','Mid Off','Mid On','Mid Wicket','Square Leg','Backward Square','Leg Gully','Leg Slip','Fine Leg','Long Leg','Long Off','Long On','Short Leg']
  shot_delivery=['straight','on','off','big shot','pull','hook','leg glance','bye area','late cut','square cut','cover']
  ball_delivery=['YORKER','FULL PITCHED','GOOD','SHORT','BOUNCER']
  bowling_styles=['Bouncer','Outswinger','Inswingers','Reverse Swing','Leg and Off Cutter','Yorker','Slower Ball','Leg Break and Off Break','Googly','Leg spinner','doosra','off spinner','Top spinner','Leg and Off Spin bowling','Carrom Ball','Off Spin bowling','Slider','Leg Spinner','Arm Ball','Off Spin bowling']
  commentary_txt=commentary['Commentary']
  dataframe=pd.DataFrame(columns=['shot','ball field','shot delivery','ball delivery','bowling style','comments'])
  batting_team=commentary_csv['Batting_Team_name']
  batsmen=commentary_csv['Batsman_name']
  bowling_team=commentary_csv['Bowling_Team_name']
  bowler=commentary_csv['Bowler_name']
  play=commentary_csv['PlayType_description']
  runs=commentary_csv['Total_Runs_on_delivery']
  SPECIAL_TOKENS  = { "bos_token": "<|BOS|>",
                    "eos_token": "<|EOS|>",
                    "unk_token": "<|UNK|>",                    
                    "pad_token": "<|PAD|>",
                    "sep_token": "<|SEP|>"}
  final=[]
  for i in range(len(commentary_csv)):
    bat_team=batting_team[i]
    batsmen_name=batsmen[i]
    bowl_team=bowling_team[i]
    bowler_name=bowler[i]
    play_type=play[i]
    run=num2words(int(runs[i]))
    comments=commentary_txt[i]
    shot=[]
    ball_f=[]
    shot_d=[]
    ball_d=[]
    bowling_s=[]
    final_txt=SPECIAL_TOKENS['bos_token']+bat_team.lower()+' '+batsmen_name.lower()+' '+bowl_team.lower()+' '+bowler_name.lower()+' '+play_type.lower()+' '+run+' '
    for word in Shots_name:
      if re.search(word.lower(),str(comments).lower()):
        final_txt=final_txt+word.lower()+' '
    for word in Ball_field:
      if re.search(word.lower(),str(comments).lower()):
        final_txt=final_txt+word.lower()+' '
    for word in shot_delivery:
      if re.search(word.lower(),str(comments).lower()):
        final_txt=final_txt+word.lower()+' '
    for word in ball_delivery:
      if re.search(word.lower(),str(comments).lower()):
        final_txt=final_txt+word.lower()+' '
    for word in bowling_styles:
      if re.search(word.lower(),str(comments).lower()):
        final_txt=final_txt+word.lower()+' '
    final_txt=final_txt+SPECIAL_TOKENS['sep_token']+str(comments).lower()+SPECIAL_TOKENS['eos_token']
    final.append(final_txt)
  return final
```
